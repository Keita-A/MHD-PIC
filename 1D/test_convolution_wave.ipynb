```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import cv2
```

```python
test_wave = np.sin(np.arange(0.0, 6.0 * np.pi, 0.1))
window_size = 5

for i in range(51):
    test_wave = np.convolve(test_wave, np.ones(window_size) / window_size, mode="same")

    if i % 10 == 0:
        plt.plot(test_wave, label=f"{i} times")

plt.legend(loc='upper right')
plt.show()
```
