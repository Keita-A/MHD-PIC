```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import cv2
```

```python
test_wave = np.sin(np.arange(0.0, 6.0 * np.pi, 0.1))
window_size = 10

fig = plt.figure(figsize=(8, 4))
for i in range(51):
    test_wave = np.convolve(test_wave, np.ones(window_size) / window_size, mode="same")

    if i % 10 == 0:
        plt.plot(test_wave, label=f"{i} times")

plt.title(f"window size = {window_size}", fontsize=20)
plt.legend(loc='upper right')
fig.savefig(f"convolution_test_window_size={window_size}.png", dpi=200)
plt.show()
```
